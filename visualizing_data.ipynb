```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import mne
from math import ceil
```

```python
chs = {'FP1': [-0.03, 0.08],
       'FP2': [0.03, 0.08],
       'F7': [-0.073, 0.047],
       'F3': [-0.04, 0.041],
       'Fz': [0, 0.038],
       'F4': [0.04, 0.041],
       'F8': [0.073, 0.047],
       'T3': [-0.085, 0],
       'C3': [-0.045, 0],
       'Cz': [0, 0],
       'C4': [0.045, 0],
       'T4': [0.085, 0],
       'T5': [-0.073, -0.047],
       'P3': [-0.04, -0.041],
       'Pz': [0, -0.038],
       'P4': [0.04, -0.041],
       'T6': [0.07, -0.047],
       'O1': [-0.03, -0.08],
       'O2': [0.03, -0.08]}
channels = pd.DataFrame(chs).transpose()
channels
```

```python
# make_dig_montage accepts only 3D coordinates
for key in chs.keys():
    chs[key]+=[0]
chs
```

```python
mont = mne.channels.make_dig_montage(chs)
mont.plot()
plt.show()
```

```python
def plot_eeg(levels, positions, axes, fig, ch_names=None, cmap='Spectral_r', cb_pos=(0.9, 0.1), cb_width=0.04, cb_height=0.9, marker=None, marker_style=None, **kwargs):
    if 'mask' not in kwargs:
        mask = np.ones(levels.shape[0], dtype='bool')
    else:
        mask = None
    im, cm = mne.viz.plot_topomap(levels, positions, axes=axes, names=ch_names,
                          cmap=cmap, mask=mask, mask_params=marker_style, show=False, **kwargs)

    cbar_ax = fig.add_axes([cb_pos[0], cb_pos[1], cb_width, cb_height])
    clb = axes.figure.colorbar(im, cax=cbar_ax)
    return im, cm
```

```python
df = pd.read_csv("eeg_data/eeg_dataset.csv")
df
```

```python
mis = df.isna().sum()
sep_col = mis[mis == df.shape[0]].index[0]
df = df.loc[:, 'main.disorder':sep_col].drop(sep_col, axis=1)
df = df.ffill()
df
```

```python
def reformat_name(name):
    '''
    reformat from XX.X.band.x.channel to band.channel
    '''
    band, _, channel = name[5:].split(sep='.')
    return f'{band}.{channel}'
reformat_vect = np.vectorize(reformat_name)
new_colnames = np.concatenate((df.columns[:2],
                              reformat_vect(df.columns[2:])))
df = df.set_axis(new_colnames, axis=1
                 )
df.columns
```

```python
numeric_cols = df.select_dtypes(include='float64').columns
print(numeric_cols)
```

```python
# mean powers per main disorder
main_mean = df.groupby('main.disorder')[numeric_cols].mean().reset_index()
# mean powers per specific disorder
spec_mean = df.groupby('specific.disorder')[numeric_cols].mean().reset_index()
# list of bands
bands = ['delta', 'theta', 'alpha', 'beta', 'highbeta', 'gamma']
# convert from wide to long
main_mean = pd.wide_to_long(main_mean, bands, ['main.disorder'], 'channel', sep='.', suffix=r'\w+')
spec_mean = pd.wide_to_long(spec_mean, bands, ['specific.disorder'], 'channel', sep='.', suffix=r'\w+')
spec_mean
```

```python
spec_mean.loc['Posttraumatic stress disorder', 'beta']
```

```python
# extract power for one main.disorder and one band
test =  spec_mean.loc['Posttraumatic stress disorder', 'beta']
# make sure that channels are in correct order
assert (test.index == channels.index).all()
# plot
fig, ax = plt.subplots()
plot_eeg(test, channels.to_numpy(), ax, fig, marker_style={'markersize':4, 'markerfacecolor':'black'})
plt.title("PTSD beta frequency")
plt.show()
```

```python
fig = plt.figure(constrained_layout=True, figsize=(12, 8))
test2 = [
     spec_mean.loc['Posttraumatic stress disorder', 'alpha'],
     spec_mean.loc['Posttraumatic stress disorder', 'beta'],
     spec_mean.loc['Posttraumatic stress disorder', 'gamma'],
     spec_mean.loc['Posttraumatic stress disorder', 'delta'],
     spec_mean.loc['Posttraumatic stress disorder', 'theta'],
     spec_mean.loc['Posttraumatic stress disorder', 'highbeta']
]
subfigs = fig.subfigures(3, 3)
for subfig, test1 in zip(subfigs.flatten(), test2):
    ax = subfig.subplots()
    plot_eeg(test1, channels.to_numpy(), ax, subfig, marker_style={'markersize':2, 'markerfacecolor':'black'})
subfigs[0, 0].suptitle('PTSD', y=1.3)
subfigs[0, 1].suptitle('', y=1.3)
subfigs[0, 2].suptitle('', y=1.3)
subfigs[0, 0].axes[0].set_ylabel('alpha')
subfigs[0, 1].axes[0].set_ylabel('beta')
subfigs[0, 2].axes[0].set_ylabel('gamma')
subfigs[1, 0].axes[0].set_ylabel('delta')
subfigs[1, 1].axes[0].set_ylabel('theta')
subfigs[1, 2].axes[0].set_ylabel('highbeta')
plt.show()

plt.show()
```
